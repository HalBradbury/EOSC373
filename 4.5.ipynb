In this exercise, you will calculate the speciation of dissolved inorganic carbon (i.e. the
concentrations of $H_2CO_3^*$, $HCO_3^-$ and $CO_3^{2-}$) from $DIC$ and $CA$. This is a calculation that
we will use extensively in the following lectures to explain how different oceanic processes
affect atmospheric $CO_2$. We will not go through the details of the calculation in class, but
after you have done this exercise, you will know how we get the numbers shown in class.
Use the following equations discussed in topic 3.8 (review topic 3.8 if needed):
$$CA = [HCO_3^-] + 2[CO_3^{2-}] \\
DIC = [H_2CO_3^*] + [HCO_3^-] + [CO_3^{2-}] \\
K_0 = \frac{[H_2CO_3^*]}{pCO_2} = 3.05 \times 10^4 \mathrm{\mu mol/kg \cdot atm}\text{ } (\text{at }T = 20\mathrm{°C}; S = 34\mathrm{‰}) \\
K_1 = \frac{[H^+][HCO_3^-]}{[H_2CO_3^*]} = 1.26 \mathrm{\mu mol/kg}\text{ } (\text{at }T = 20\mathrm{°C}; S = 34\mathrm{‰}) \\
K_2 = \frac{[H^+][CO_3^{2-}]}{[HCO_3^-]} = 1.0 \times 10^{-3} \mathrm{\mu mol/kg }\text{ } (\text{at }T = 20\mathrm{°C}; S = 34\mathrm{‰})$$

Measuring $DIC$ and $CA$ $(\mathrm{\mu mol/kg})$, and knowing $K$’s, provides 5 equations for 5 unknowns.
Solving these equation gives:

$$[CO_3^{2-}] = \frac{(CA \times K_r)-(DIC \times K_r)-(4 \times CA) + \sqrt{\xi}}{2(K_r-4)} \text{ } (\mathrm{\mu mol/kg})$$
Where: $K_r = \frac{K_1}{K_2}$
$$\xi = ((4 \times CA) + (DIC \times K_r) – (CA \times K_r))^2 + (4 \times (K_r - 4) \times CA^2) (\mathrm{\mu mol/kg})^2$$
